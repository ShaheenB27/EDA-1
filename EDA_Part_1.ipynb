# **Data Type Conversion**

```python
data['Month']=pd.to_numeric(data['Month'],errors='coerce')
data['Temp C']=pd.to_numeric(data['Temp C'],errors='coerce') #Coerce will intoduce NA values for non numeric data in the column
data['Weather']=data['Weather'].astype('category') #data[Wind]=data[Wind].astype('int64')
```

```python
data_cleaned1=data.drop_duplicates()
```

```python
data_cleaned1.shape
```

# **Drop Columns**

```python
data_cleaned2=data_cleaned1.drop('Temp C',axis=1)
```

```python
data_cleaned2
```

# **Rename the Columns**

```python
#rename the Solar column
data_cleaned2 = data_cleaned2.rename({'Solar.R': 'Solar'}, axis=1)
```

```python
data_cleaned2
```

# **Missing Values and Imputation**

```python
import seaborn as sns
cols = data_cleaned2.columns
colours = ['#000099','#ffff00'] #specify the colours yellow is -missing blue is - not missing.
sns.heatmap(data_cleaned2[cols].isnull(),
cmap=sns.color_palette(colours))
```

```python
data_cleaned2[data_cleaned2.isnull().any(axis=1)].head()
```

```python
data_cleaned2.isnull().sum()
```

```python
#Median Imputation
median = data_cleaned2['Ozone'].median()
print(median)
```

```python
data_cleaned2['Ozone'] = data_cleaned2['Ozone'].fillna(median)
```

```python
data_cleaned2
```

```python
data_cleaned2.boxplot(figsize=(15,15))
```

```python
#Missing value imputation for categorical vlaue
#Get the object columns
obj_columns=data_cleaned2[['Weather']]
```

```python
obj_columns.isnull().sum()
```

```python
#Missing value imputation for categorical vlaue
obj_columns=obj_columns.fillna(obj_columns.mode().iloc[0])
```

```python
obj_columns.isnull().sum()
```
